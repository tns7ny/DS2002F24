# **Exercise 2: Sorting Results**

## 1.

```python
# Write a query to find all movies where the genre is "Comedy" and sort them by IMDb
# rating in descending order (Return the first 5 results).

comedy_movies_by_rating = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)

for movie in comedy_movies_by_rating:
    pprint.pprint(movie)
```

## 2.

```python
# Write a query to find all movies where the genre is "Drama" and sort them by release year
# in ascending order (Return the first 5 results).

drama_movies_by_yr = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)

for movie in drama_movies_by_yr:
    pprint.pprint(movie)
```

# **Exercise 3: Aggregation Pipeline**



## 1.

```python
# Write an aggregation pipeline that calculates the average IMDb rating for each genre (Return the top 5 genres).

avg_rating_by_genre = collection.aggregate([
{"$unwind": "$genres"},
{"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
{"$sort": {"avg_rating": -1}},
{"$limit": 5}
])

for group in avg_rating_by_genre:
    pprint.pprint(group)
```

## 2.

```python
# Write an aggregation pipeline to find the top 5 directors by the average IMDb rating of their movies.

avg_rating_top_directors = collection.aggregate([
{"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
{"$sort": {"avg_rating": -1}},
{"$limit": 5}
])

for group in avg_rating_top_directors:
    pprint.pprint(group)
```

## 3.

```python
# Write an aggregation pipeline to calculate the total number of movies released in each
# year (Sort the results by the year).

total_movies_by_year = collection.aggregate([
{"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
{"$sort": {"_id": 1}}
])

for group in total_movies_by_year:
    pprint.pprint(group)
```

# **Exercise 4: Updating and Deleting Documents**



## 1.

```python
# Write a query to update the IMDb rating of a movie with the title "The Godfather" to 9.5.

collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})

for movie in collection.find({"title": "The Godfather"}):
    pprint.pprint(movie)
```

## 2.

```python
# Write a query to update all movies where the genre is "Horror" and set their IMDb rating
# to 6.0 if it is currently null.

collection.update_many(
    {"genres": "Horror", "imdb.rating": {"$exists": False}},
    {"$set": {"imdb.rating": 6.0}}
)

# Check
updated_horror_movies = collection.find({"genres": "Horror", "imdb.rating": 6.0}).limit(5)

for movie in updated_horror_movies:
    pprint.pprint(movie)
```

## 3.

```python
# Write a query to delete all movies that were released before the year 1950.

collection.delete_many({"year": {"$lt": 1950}})

# Check
# Should not return anything
deleted_movies = collection.find({"year": {"$lt": 1950}})

for movie in deleted_movies:
    pprint.pprint(movie)
```
