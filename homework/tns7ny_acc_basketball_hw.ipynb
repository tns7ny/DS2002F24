```python
import pandas as pd

# 1A
# Load data
acc_players_df = pd.read_csv('/content/acc_players-2324F.csv', header=1)

# 1B
# Display first 5 rows
print("ACC 2023-2024 Player Stats:")
acc_players_df.head()
```

```python
# 2A
# Calculate the total points scored by all players combined
total_points = acc_players_df['PTS'].sum()
print(f"Total Points Scored by All Players: {total_points}")
```

```python
# 2B
# Find the name and their stats for the player who has played the most minutes (MP)
most_mins_player = acc_players_df.loc[acc_players_df['MP'].idxmax()]
print(f"Player with the Most Minutes Played: {most_mins_player['Player']} ({most_mins_player['MP']} minutes)")
most_mins_player
```

```python
# 2C
# Identify the top 5 players in terms of total rebounds (TRB)
top_rebounds_players = acc_players_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 Players by Total Rebounds:")
top_rebounds_players
```

```python
# 3A
# Create a new DataFrame containing only players who played more than 500 minutes
players_over_500_mins = acc_players_df[acc_players_df['MP'] > 500]
print("Players who Played Over 500 Minutes:")
players_over_500_mins.head()
```

```python
# 3B
# From this filtered DataFrame, determine the player with the highest total assists (AST)
highest_assists_player = players_over_500_mins.loc[players_over_500_mins['AST'].idxmax()]
print(f"Player with the Highest Total Assists")
highest_assists_player
```

```python
# 3C
# Who are the top 3 Assist Leads in the League?
top_assist_leads = acc_players_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print("Top 3 Assist Leads in the ACC:")
top_assist_leads

# This gives same output, but uses players_over_500_mins dataframe:
# top_assist_leads2 = players_over_500_mins[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
# print("Top 3 Assist Leads in the ACC:")
# top_assist_leads2
```

```python
# 3D
# Who are the top 3 Shot Blockers?
top_shot_blockers = acc_players_df[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
print("Top 3 Shot Blockers in the ACC:")
top_shot_blockers

# This gives same output, but uses players_over_500_mins dataframe:
# top_shot_blockers2 = players_over_500_mins[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
# print("Top 3 Shot Blockers in the ACC:")
# top_shot_blockers2
```

```python
# 4A
# Group the players by School and calculate the total points scored by each school
school_points = acc_players_df.groupby('School')[['PTS']].sum().reset_index()
print("Total Points Scored by School:")
school_points.head()
```

```python
# 4B
# Group the players by School and calculate the total assists (AST) for each team
school_assists = acc_players_df.groupby('School')[['AST']].sum().reset_index()
print("Total Assists by School:")
school_assists.head()
```

```python
# 4C
# Sort the schools by total points scored and display the top 3 schools
top_3_schools = school_points.sort_values(by='PTS', ascending=False)
print("Top 3 Schools by Total Points Scored:")
top_3_schools.head(3)
```

```python
# 5A Bonus
# Create a bar chart showing the top 5 players by total points scored (PTS)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

top_5_players_PTS = acc_players_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='PTS', y='Player', data=top_5_players_PTS)
plt.title('Top 5 Players by Total Points Scored (PTS)')
plt.xlabel('Total Points Scored (PTS)')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```

```python
# 5B Bonuns
# Correlation: Investigate whether there is any correlation between a player's field goal
# percentage (FG%) and their win shares (WS). Provide a scatter plot and calculate the
# correlation coefficient

correlation_coefficient = acc_players_df['FG%'].corr(acc_players_df['WS'])
print(f"Correlation Coefficient between FG% and WS: {correlation_coefficient}")
```

```python
# 5B Bonus (continued)

plt.figure(figsize=(10, 6))
plt.scatter(acc_players_df['FG%'], acc_players_df['WS'])
plt.title('Relationship between FG% and WS')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.grid(True)
plt.show()
```
